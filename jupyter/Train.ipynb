```python
import pandas as pd

df = pd.read_excel('train.xlsx')
```

```python
df['bdate'].value_counts()
df['bdate'] = pd.to_datetime(df['bdate'])
```

```python
df
```

```python
from datetime import date,datetime

today= date.today()

year= []

for i in df['bdate']:
    year.append(i.year)

df1 = df.assign(year=year)

df1
```

```python
age = []

for i in df1['year']:
    age.append(2022-i)

df1.drop(['bdate', 'year'], axis=1, inplace=True)

df1 = df1.assign(age=age)

df1
```

```python
df1.info()

mn = int(df1['age'].mean())

print(mn)
```

```python
df1['age'].replace(0,mn,inplace=True)

df1
```

```python
df1[list(pd.get_dummies(df1['education_form']).columns)] = pd.get_dummies(df1["education_form"])
```

```python
df1.drop('education_form', axis=1, inplace=True)

df1
```

```python
df1[list(pd.get_dummies(df1['education_status']).columns)] = pd.get_dummies(df1['education_status'])

df1.drop('education_status', axis = 1, inplace =True)
```

```python
df1
```

```python
df1[list(pd.get_dummies(df1['occupation_type']).columns)] = pd.get_dummies(df1['occupation_type'])

df1.drop('occupation_type', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
import xgboost
from sklearn.metrics import confusion_matrix, accuracy_score

df1.drop(['Part-time', 'Candidate of Sciences',"Student (Bachelor's)","Student (Master's)"], axis=1, inplace=True)

corr = df1.corr()
```

```python
df1.drop(['has_photo','has_mobile','graduation','life_main','people_main','Distance Learning','PhD','Student (Specialist)','Undergraduate applicant','work'],axis=1, inplace=True)

corr = df1.corr()

corr.style.background_gradient(cmap='coolwarm')
```

```python
df1.drop(['city','career_start','career_end'],axis=1,inplace=True)
df1.info()
```

```python
x = df1.drop(["result"],axis=1)

y = df1['result']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)

XGB = xgboost.XGBClassifier()

XGB.fit(x_train, y_train)
```

```python
y_pred = XGB.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```
